```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/Regression-AreaPrice (1) - Regression-AreaPrice (1).csv')
print(df)
```

```python
print("Defining X and Y Variable")
x=df['Area']
y=df['Price']
plt.scatter(x,y,color='blue')
```

```python
print("Splitting Data into train and test")
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
plt.scatter(x_train,y_train,color='blue')
plt.scatter(x_test,y_test,color='red')
print("Training the model")
```

```python
model=LinearRegression()
model.fit(x_train.values.reshape(-1,1),y_train)
```

```python
print("Predicting using the trained model-x_test")
y_pred=model.predict(x_test.values.reshape(-1,1))
print(y_test)
print(y_pred)
plt.scatter(x_train,y_train,color='blue')
plt.scatter(x_test,y_pred,color='red')
plt.xticks()
plt.yticks()
plt.show()
```

```python
print("Finding intercept & coefficient")
print(model.intercept_)
print(model.coef_)
print(model.coef_,'x+',model.intercept_)
```
