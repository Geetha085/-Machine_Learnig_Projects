```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
x=[[150],[160],[170],[180],[190],[200],[210],[220]]
y=[0,0,0,1,1,1,1, 1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
clf=DecisionTreeClassifier(max_depth=1)
clf.fit(x_train,y_train)
```

```python
plt.figure(figsize=(8,6))
tree.plot_tree(clf,filled=True,
               feature_names=['Weight'],
               class_names=['Apple','Orange'],
               rounded=True,
               fontsize=12)
```

```python
y_pred=clf.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuracy of the univariate decision tree: {accuracy:.2f}")
new_fruit_weight=[[175]]
prediction=clf.predict(new_fruit_weight)
print("Prediction (0=Apple,1=Orange):",prediction[0])
```
