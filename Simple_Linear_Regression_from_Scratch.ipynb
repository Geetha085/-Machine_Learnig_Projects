```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours_studied = np.array([1, 2, 3, 4, 5])
scores = np.array([15, 30, 45, 60, 75])
```

```python
class SimpleLinearRegression:
    def __init__(self):
        self.slope = 0
        self.intercept = 0

    def fit(self, X, y):
        n = len(X)
        x_mean = np.mean(X)
        y_mean = np.mean(y)

        # Calculate the slope (m)
        self.slope = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean) ** 2)

        # Calculate the intercept (b)
        self.intercept = y_mean - (self.slope * x_mean)

    def predict(self, X):
        return self.slope * X + self.intercept
```

```python
model = SimpleLinearRegression()
model.fit(hours_studied, scores)
```

```python

predicted_scores = model.predict(hours_studied)
```

```python
plt.scatter(hours_studied, scores, color='blue', label='Actual Scores')
plt.plot(hours_studied, predicted_scores, color='red', label='Fitted Line')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Simple Linear Regression from Scratch')
plt.legend()
plt.show()
```
